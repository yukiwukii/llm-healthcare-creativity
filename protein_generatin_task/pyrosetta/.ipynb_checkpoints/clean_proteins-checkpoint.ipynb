```python
import pyrosetta
pyrosetta.init()
```

```python
# Load OmpT structure
ompT_pose = pyrosetta.pose_from_pdb("1i78.pdb")

# Load Penetratin structure (Your PDB file)
penetratin_pose = pyrosetta.pose_from_pdb("penetratin_single.pdb")

print(f"OmpT loaded with {ompT_pose.total_residue()} residues.")
print(f"Penetratin loaded with {penetratin_pose.total_residue()} residues.")
```

```python
for i in range(1, penetratin_pose.total_residue() + 1):
    res_name = penetratin_pose.residue(i).name()
    chain_id = penetratin_pose.pdb_info().chain(i)
    print(f"Residue {i}: {res_name} (Chain {chain_id})")
```

```python
# Load the full 1OMQ PDB file
full_pose = pyrosetta.pose_from_pdb("1omq.pdb")

# Identify the correct chain (Assuming Chain A contains a single Penetratin)
chain_to_keep = "A"
extracted_pose = pyrosetta.Pose()

# Add the first 16 residues from Chain A only
residue_count = 0
for i in range(1, full_pose.total_residue() + 1):
    if full_pose.pdb_info().chain(i) == chain_to_keep:
        extracted_pose.append_residue_by_jump(full_pose.residue(i), max(1, residue_count))
        residue_count += 1
        if residue_count == 16:
            break  # Stop after extracting 16 residues

# Save the cleaned-up PDB file with only one copy of Penetratin
extracted_pose.dump_pdb("penetratin_single.pdb")

print(f"✅ Extracted {extracted_pose.total_residue()} residues and saved as 'penetratin_single.pdb'.")
```
